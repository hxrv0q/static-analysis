## Завдання 2

Використовуючи набір даних TitanicDataset змінити пропущені значення віку на середнє, в залежності від статі і класу каюти.

Спочатку завантажуємо набір даних. Після цього я заміню пропущені значення віку на середнє значення віку в залежності від статі та класу каюти. Та вивожу оновлений набір даних 

```python
titanic_data = pd.read_csv("TitanicDataset.csv")

mean_ages_by_sex_places = titanic_data.groupby(["Sex", "Pclass"])["Age"].mean()

for (sex, pclass), mean_age in mean_ages_by_sex_places.items():
    titanic_data.loc[
        (titanic_data["Age"].isnull())
        & (titanic_data["Sex"] == sex)
        & (titanic_data["Pclass"] == pclass),
        "Age",
    ] = mean_age

titanic_data
```

Як бачимо 2017 року немає у наборі даних, відповідно немає даних про жоден злочин, який стався після 12 години дня та до 6 години ранку кожного понеділка 2017 року. Спробуємо тоді для 2016 року.

```python
crimes_df = pd.read_excel("CrimesData.xlsx", sheet_name="2016")
crimes_df["DateTime"] = pd.to_datetime(
    crimes_df["CrimeDate"] + " " + crimes_df["CrimeTime"],
    format="mixed",
    errors="coerce",
)
filtered_crimes = crimes_df[
    (crimes_df["DateTime"].dt.year == 2016)
    & (crimes_df["DateTime"].dt.dayofweek == 0)
    & ((crimes_df["DateTime"].dt.hour > 12) | (crimes_df["DateTime"].dt.hour < 6))
]

filtered_crimes
```

Виведемо кількість злочинів, що сталися після 12 години і до 6 ранку кожного понеділка 2016 року.
